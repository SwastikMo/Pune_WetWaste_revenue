```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("C:\\Users\\swast\\Downloads\\Solid Waste Management Revenue.csv")
```

```python
df
```

```python
df=df.dropna()
```

```python
x=df.iloc[:,[1,9,16]]
print(x)
x1s=x.iloc[:,1].values.reshape(-1,1)
x2s=x.iloc[:,2].values.reshape(-1,1)
x1=list(x.iloc[:,1])
x2=list(x.iloc[:,2])
y=list(x.iloc[:,0])
```

```python
bar1=np.arange(len(x))
```

```python
plt.bar(y,x1,-0.25,color="red",align='edge')
plt.bar(y,x2,0.25,color="navy",align='edge')
plt.xlabel('year')
plt.ylabel('Annual Expenditure( red ) and \n Revenue( blue ) \n in lakhs')
```

```python
'''  we could observe the increase in annual expenditure in handling wastes but are observing a much lower amount of revenue
     generated due to outdated handling and lack of proper infrastructure to handle the waste generated and convert it into 
     valuable products.                                                                                                      '''
```

```python
ln=LinearRegression()
ln2=LinearRegression()
bar1=bar1.reshape(-1,1)
```

```python
ln.fit(bar1,x1s)
ln2.fit(bar1,x2s)
```

```python
print(ln.predict([[15]]))
ln2.predict([[15]])
```

```python
# the predicted linear model says that 12 to 15 years down the line the expenditure on handling waste is going to be about 1
# hundred thousand lakhs about 2.5 times the current expenditure while the revenue seems to be on a decreasing trend predicting a
# collection of about only 19000 lakhs which is just about 10% of the total expenditure 
# showing the importance of using new waste managment and hence maximizing revenue generation 
```
